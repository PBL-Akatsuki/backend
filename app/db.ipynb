```python
import pandas as pd
import numpy as np
import random
from datetime import datetime, timedelta

# Number of logs to generate
n = 3000

# Define sci-fi categorical values
player_levels = ["Cosmic Initiate", "Galactic Cadet",
                 "Starborne Operative", "Celestial Vanguard", "Quantum Sovereign"]
player_ranks = ["Nebula Novice", "Stellar Scout", "Cosmic Ensign",
                "Interstellar Agent", "Galactic Commander", "Solar Admiral", "Universal Archon"]
teams = ["Cyber Sentinels", "Shadow Collective", "Lone Wolf"]
dark_market_types = ["RealityBender", "GlitchChip", "NeonShifter", "DataDrifter", "QuantumBreaker",
                     "CyberWraith", "PixelPirate", "GhostCoder", "ByteBandit", "Error404"]

# Prepare lists to store generated values
player_ids = []
timestamps = []
hours_played = []
money_spent = []
criminal_score = []
missions_completed = []
player_rank = []
team_affiliation = []
vip_status = []
cash_on_hand = []
sync_stability = []
quest_exploit_score = []
player_level = []
dark_market_transactions = []
transaction_amount = []
neural_link_stability = []

# Define a date range for Timestamps (e.g., from 2025-02-15 to 2025-02-18)
start_date = datetime(2025, 2, 15, 0, 0, 0)
end_date = datetime(2025, 2, 18, 0, 0, 0)
delta_seconds = int((end_date - start_date).total_seconds())

# Generate 3000 logs
for i in range(1, n + 1):
    # Player ID: e.g., "P0001", "P0002", ...
    pid = f"P{i:04d}"
    player_ids.append(pid)

    # Random Timestamp within the specified range
    random_seconds = random.randint(0, delta_seconds)
    ts = start_date + timedelta(seconds=random_seconds)
    timestamps.append(ts.strftime("%Y-%m-%d %H:%M:%S"))

    # Determine if this record is suspicious (biased toward the culprit)
    is_suspicious = random.random() < 0.15  # 15% chance

    # Hours Played:
    # Normal: 50-300 hours, Suspicious: 400-500 hours (abnormally high)
    if is_suspicious:
        hp = random.randint(400, 500)
    else:
        hp = random.randint(50, 300)
    hours_played.append(hp)

    # Money Spent ($):
    # Normal: 1,000-20,000, Suspicious: extremely low (1-50) despite high hours played
    if is_suspicious:
        ms = random.randint(1, 50)
    else:
        ms = random.randint(1000, 20000)
    money_spent.append(ms)

    # Criminal Score:
    # Normal: 0-5, Suspicious: 5-10 (higher score)
    if is_suspicious:
        cs = random.randint(5, 10)
    else:
        cs = random.randint(0, 5)
    criminal_score.append(cs)

    # Missions Completed:
    # Normal: 20-300 missions, Suspicious: 300-500 (unusually high activity)
    if is_suspicious:
        mc = random.randint(300, 500)
    else:
        mc = random.randint(20, 300)
    missions_completed.append(mc)

    # Player Rank:
    # Suspicious players are more likely to have higher ranks
    if is_suspicious:
        # choose from ranks with higher prestige
        pr = random.choice(player_ranks[3:])
    else:
        pr = random.choice(player_ranks[:4])
    player_rank.append(pr)

    # Team Affiliation:
    # Suspicious players are forced into "Shadow Collective"
    if is_suspicious:
        team = "Shadow Collective"
    else:
        team = random.choice(teams)
    team_affiliation.append(team)

    # VIP Status:
    # Normal: 10% chance to be VIP; Suspicious: 5% chance
    if is_suspicious:
        vip = "Yes" if random.random() < 0.05 else "No"
    else:
        vip = "Yes" if random.random() < 0.10 else "No"
    vip_status.append(vip)

    # Cash on Hand ($):
    # Normal: 500-15,000, Suspicious: very low (0-500)
    if is_suspicious:
        coh = random.randint(0, 500)
    else:
        coh = random.randint(500, 15000)
    cash_on_hand.append(coh)

    # Sync Stability (%):
    # Normal: 85-100%, Suspicious: lower stability (40-80%)
    if is_suspicious:
        ss = round(random.uniform(40, 80), 1)
    else:
        ss = round(random.uniform(85, 100), 1)
    sync_stability.append(ss)

    # Quest Exploit Score:
    # Normal: 300-1000, Suspicious: often 0 or low (0-300)
    if is_suspicious:
        qes = 0 if random.random() < 0.5 else random.randint(0, 300)
    else:
        qes = random.randint(300, 1000)
    quest_exploit_score.append(qes)

    # Player Level:
    # Suspicious players tend to be less experienced: "Cosmic Initiate" or "Galactic Cadet"
    if is_suspicious:
        pl = random.choice(player_levels[:2])
    else:
        pl = random.choice(player_levels[2:])
    player_level.append(pl)

    # Dark Market Transactions:
    # Suspicious players more likely use certain illicit transaction types
    if is_suspicious:
        dmt = random.choice(
            ["RealityBender", "GlitchChip", "CyberWraith", "Error404"])
    else:
        dmt = random.choice(dark_market_types)
    dark_market_transactions.append(dmt)

    # Transaction Amount ($):
    # Normal: 100-10,000, Suspicious: set to be unusually low (50-200)
    if is_suspicious:
        ta = random.randint(50, 200)
    else:
        ta = random.randint(100, 10000)
    transaction_amount.append(ta)

    # Neural Link Stability (%):
    # Normal: 60-100%, Suspicious: lower (40-70%)
    if is_suspicious:
        nls = round(random.uniform(40, 70), 1)
    else:
        nls = round(random.uniform(60, 100), 1)
    neural_link_stability.append(nls)

# Create DataFrame
data = pd.DataFrame({
    "Player ID": player_ids,
    "Timestamps": timestamps,
    "Hours Played": hours_played,
    "Money Spent ($)": money_spent,
    "Criminal Score": criminal_score,
    "Missions Completed": missions_completed,
    "Player Rank": player_rank,
    "Team Affiliation": team_affiliation,
    "VIP Status": vip_status,
    "Cash on Hand ($)": cash_on_hand,
    "Sync Stability (%)": sync_stability,
    "Quest Exploit Score": quest_exploit_score,
    "Player Level": player_level,
    "Dark Market Transactions": dark_market_transactions,
    "Transaction Amount ($)": transaction_amount,
    "Neural Link Stability (%)": neural_link_stability
})

# Save the generated data to a CSV file
data.to_csv("neoverse_logs.csv", index=False)

print("3000 logs generated and saved to 'neoverse_logs.csv'")
```

```python
# Function to detect and remove outliers using IQR
def remove_outliers(df, columns):
    for col in columns:
        Q1 = df[col].quantile(0.25)
        Q3 = df[col].quantile(0.75)
        IQR = Q3 - Q1
        lower_bound = Q1 - 1.5 * IQR
        upper_bound = Q3 + 1.5 * IQR
        df = df[(df[col] >= lower_bound) & (df[col] <= upper_bound)]
    return df
```

```python
def getMinimumValue(power, armor):
    # Try using armor in each round and find the minimum health needed
    min_health = float('inf')

    # Try each position for the armor (including not using it)
    for armor_pos in range(-1, len(power)):  # -1 means don't use armor
        # Calculate total damage and find minimum health required
        health = 1  # Player must have at least 1 health at the end

        # Go through rounds in reverse
        for i in range(len(power)-1, -1, -1):
            # Add the damage for this round
            damage = power[i]
            if i == armor_pos:
                # Using armor in this round
                damage = max(0, power[i] - armor)

            # Add damage to required health
            health += damage

        # Update minimum health if this strategy is better
        min_health = min(min_health, health)

    return min_health

# Test cases based on the examples in the images


def run_tests():
    # Test Case 1: Example from the problem statement
    power1 = [1, 2, 6, 7]
    armor1 = 5
    expected1 = 12
    result1 = getMinimumValue(power1, armor1)
    print(
        f"Test Case 1: {result1 == expected1}, Expected: {expected1}, Got: {result1}")

    # Test Case 2: From Sample Case 0
    power2 = [1, 2, 3]
    armor2 = 1
    expected2 = 6
    result2 = getMinimumValue(power2, armor2)
    print(
        f"Test Case 2: {result2 == expected2}, Expected: {expected2}, Got: {result2}")

    # Test Case 3: From Sample Case 1
    power3 = [1, 2, 3]
    armor3 = 4
    expected3 = 4
    result3 = getMinimumValue(power3, armor3)
    print(
        f"Test Case 3: {result3 == expected3}, Expected: {expected3}, Got: {result3}")

    # Test Case 4: Edge case with one round
    power4 = [5]
    armor4 = 3
    expected4 = 3  # 5-3 = 2 damage plus 1 health at end
    result4 = getMinimumValue(power4, armor4)
    print(
        f"Test Case 4: {result4 == expected4}, Expected: {expected4}, Got: {result4}")

    # Test Case 5: Edge case where armor completely blocks
    power5 = [2, 4, 1]
    armor5 = 4
    expected5 = 4  # 2+1+1 = 4 (using armor on round 2)
    result5 = getMinimumValue(power5, armor5)
    print(
        f"Test Case 5: {result5 == expected5}, Expected: {expected5}, Got: {result5}")


# Run the tests
if __name__ == "__main__":
    run_tests()
```

```python
def analyzeProductCode(productSeq):
    n = len(productSeq)
    total_valid = 0

    for start in range(n):
        # Use an array for faster lookups (only 7 characters 'a' to 'g')
        char_freq = [0] * 7  # indexed 0-6 for 'a'-'g'
        max_freq = 0
        distinct_chars = 0

        for end in range(start, n):
            idx = ord(productSeq[end]) - ord('a')

            # If this is the first occurrence of this character
            if char_freq[idx] == 0:
                distinct_chars += 1

            # Increment frequency
            char_freq[idx] += 1

            # Update max frequency if needed
            if char_freq[idx] > max_freq:
                max_freq = char_freq[idx]

            # Check validity with a simple comparison
            if max_freq <= distinct_chars:
                total_valid += 1

    return total_valid


def test_analyze_product_code():
    test_cases = [
        ("abab", 10),
        ("aacd", 9),
        ("abaa", 8),
        ("a", 1),
        ("aaa", 3),
        ("abc", 6),
        ("abcabc", 21)
    ]

    for i, (test_input, expected) in enumerate(test_cases, 1):
        result = analyzeProductCode(test_input)
        status = "PASS" if result == expected else "FAIL"
        print(
            f"Test {i}: {test_input}  Expected: {expected}, Got: {result} - {status}")


# Run tests
test_analyze_product_code()
```

```python
def count_valid_substrings(productSeq):
    n = len(productSeq)
    result = 0
    max_length = 26 * 26  # Maximum possible valid substring length

    for i in range(n):
        counts = [0] * 26
        max_freq = 0
        distinct = 0

        for j in range(i, max(i - max_length, -1), -1):
            char_code = ord(productSeq[j]) - ord('a')

            if counts[char_code] == 0:
                distinct += 1
            counts[char_code] += 1

            if counts[char_code] > max_freq:
                max_freq = counts[char_code]

            if max_freq <= distinct:
                result += 1
            else:
                break  # Further extensions to the left are invalid

    return result
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import LabelEncoder, MinMaxScaler
from sklearn.ensemble import RandomForestRegressor
import warnings
warnings.filterwarnings('ignore')

# For reproducibility
np.random.seed(42)

# Load the data
# If loading from a CSV file:
data = pd.read_csv('updated_neoverse_logs.csv')

# 1. Data Preprocessing
# Convert timestamp to datetime
data['Timestamps'] = pd.to_datetime(data['Timestamps'])

# Extract datetime features (optional, but often useful)
data['Day'] = data['Timestamps'].dt.day
data['Month'] = data['Timestamps'].dt.month
data['Hour'] = data['Timestamps'].dt.hour
data['Weekday'] = data['Timestamps'].dt.weekday

# Drop the original timestamp column as it's now encoded
data = data.drop('Timestamps', axis=1)

# Identify categorical and numerical columns
categorical_columns = ['Player ID', 'Player Rank', 'Team Affiliation', 'VIP Status',
                       'Player Level', 'Dark Market Transactions']
numerical_columns = ['Hours Played', 'Money Spent ($)', 'Criminal Score', 'Missions Completed',
                     'Cash on Hand ($)', 'Sync Stability (%)', 'Quest Exploit Score',
                     'Transaction Amount ($)', 'Neural Link Stability (%)',
                     'Day', 'Month', 'Hour', 'Weekday']

# 2. Label Encoding for categorical features
# Initialize a dictionary to store the encoders
label_encoders = {}
for column in categorical_columns:
    le = LabelEncoder()
    data[f'{column}_encoded'] = le.fit_transform(data[column])
    label_encoders[column] = le

# Store original data before dropping columns (for visualization comparison)
data_with_categories = data.copy()

# Drop original categorical columns as requested
data = data.drop(categorical_columns, axis=1)

# 3. Visualize boxplots for outlier detection - BEFORE any removal


def plot_detailed_boxplots(data, columns, title_prefix="", output_file='outlier_boxplots.png'):
    """
    Create detailed horizontal boxplots with outlier information.

    Parameters:
    - data: DataFrame containing the data
    - columns: List of column names to visualize
    - title_prefix: Optional prefix for plot title
    - output_file: Path to save the output figure
    """
    n_cols = min(3, len(columns))
    n_rows = (len(columns) // n_cols) + (1 if len(columns) % n_cols > 0 else 0)

    plt.figure(figsize=(20, n_rows * 5))

    for i, column in enumerate(columns):
        plt.subplot(n_rows, n_cols, i+1)

        # Calculate IQR for outlier detection
        Q1 = data[column].quantile(0.25)
        Q3 = data[column].quantile(0.75)
        IQR = Q3 - Q1
        lower_bound = Q1 - 1.5 * IQR
        upper_bound = Q3 + 1.5 * IQR

        # Create enhanced boxplot
        sns.boxplot(x=data[column], orient='h', color='skyblue', width=0.5)

        # Add vertical lines for IQR boundaries
        plt.axvline(x=lower_bound, color='red', linestyle='--', linewidth=1.5,
                    label=f'Lower bound: {lower_bound:.2f}')
        plt.axvline(x=upper_bound, color='red', linestyle='--', linewidth=1.5,
                    label=f'Upper bound: {upper_bound:.2f}')

        # Add distribution plot above the boxplot for better visualization
        ax2 = plt.gca().twiny()
        sns.kdeplot(data[column], color='navy', alpha=0.6, ax=ax2)
        ax2.set_yticks([])
        ax2.set_ylabel('')

        # Count outliers
        outlier_mask = (data[column] < lower_bound) | (
            data[column] > upper_bound)
        outlier_count = outlier_mask.sum()
        outlier_percent = (outlier_count / len(data)) * 100

        # Add more detailed title with outlier information
        plt.title(f"{title_prefix} {column}\n"
                  f"Outliers: {outlier_count} ({outlier_percent:.2f}%)\n"
                  f"IQR: {IQR:.2f}  |  Q1: {Q1:.2f}  |  Q3: {Q3:.2f}",
                  fontsize=12)

        # Add labels with more statistics
        plt.xlabel(f'Min: {data[column].min():.2f} | Max: {data[column].max():.2f} | Mean: {data[column].mean():.2f} | Median: {data[column].median():.2f}',
                   fontsize=10)

        # If there are outliers, highlight them on the plot
        if outlier_count > 0:
            outlier_values = data[column][outlier_mask]
            plt.plot(outlier_values, np.zeros_like(outlier_values), 'ro', alpha=0.5,
                     markersize=5, label=f'Outliers ({outlier_count})')

        plt.legend(loc='best', fontsize=9)
        plt.grid(True, linestyle='--', alpha=0.7)

    plt.tight_layout()
    plt.savefig(output_file, dpi=300, bbox_inches='tight')
    plt.close()


# Create boxplots for outlier visualization before removal
plot_detailed_boxplots(data, numerical_columns,
                       title_prefix="Boxplot for",
                       output_file='outliers_before_removal.png')

# 4. Outlier removal using IQR method


def remove_outliers(df, columns, plot_before_after=True):
    """
    Removes outliers from specified columns using the IQR method.

    Parameters:
    - df: DataFrame containing the data
    - columns: List of column names to check for outliers
    - plot_before_after: Whether to generate before/after distribution plots

    Returns:
    - df_clean: DataFrame with outliers removed
    - outlier_summary: DataFrame with outlier statistics by column
    """
    df_clean = df.copy()
    outlier_indices = []
    outlier_summary = []

    # Track outliers by column
    for column in columns:
        # Calculate IQR
        Q1 = df[column].quantile(0.25)
        Q3 = df[column].quantile(0.75)
        IQR = Q3 - Q1

        # Define bounds
        lower_bound = Q1 - 1.5 * IQR
        upper_bound = Q3 + 1.5 * IQR

        # Find outliers
        column_outliers = df[(df[column] < lower_bound) |
                             (df[column] > upper_bound)].index
        outlier_count = len(column_outliers)
        outlier_percent = (outlier_count / len(df)) * 100

        # Add to summary
        outlier_summary.append({
            'Column': column,
            'Total_Count': len(df),
            'Outlier_Count': outlier_count,
            'Outlier_Percentage': outlier_percent,
            'Min_Value': df[column].min(),
            'Max_Value': df[column].max(),
            'Lower_Bound': lower_bound,
            'Upper_Bound': upper_bound
        })

        # Add to overall outlier indices
        outlier_indices.extend(column_outliers)

    # Get unique indices of all outliers
    outlier_indices = list(set(outlier_indices))

    # Remove outliers
    df_clean = df_clean.drop(outlier_indices)

    # Create summary DataFrame
    outlier_summary_df = pd.DataFrame(outlier_summary)

    # Print summary
    print(f"Total rows before outlier removal: {len(df)}")
    print(
        f"Identified {len(outlier_indices)} unique rows containing outliers ({len(outlier_indices)/len(df)*100:.2f}%)")
    print(f"Rows remaining after outlier removal: {len(df_clean)}")

    # Optional: Plot distributions before and after outlier removal
    if plot_before_after and len(columns) > 0:
        n_cols = min(3, len(columns))
        n_rows = (len(columns) // n_cols) + \
            (1 if len(columns) % n_cols > 0 else 0)

        plt.figure(figsize=(n_cols*7, n_rows*5))

        for i, column in enumerate(columns):
            plt.subplot(n_rows, n_cols, i+1)

            # Original distribution with outliers
            sns.kdeplot(df[column], color='red', label='With Outliers')

            # New distribution without outliers
            sns.kdeplot(df_clean[column], color='blue',
                        label='Without Outliers')

            # Calculate stats for both distributions
            mean_before = df[column].mean()
            mean_after = df_clean[column].mean()
            std_before = df[column].std()
            std_after = df_clean[column].std()

            plt.title(f'Distribution of {column} Before/After Outlier Removal')
            plt.xlabel(
                f'{column} Value\nBefore: ={mean_before:.2f}, ={std_before:.2f}\nAfter: ={mean_after:.2f}, ={std_after:.2f}')
            plt.ylabel('Density')
            plt.legend()
            plt.grid(True, linestyle='--', alpha=0.6)

        plt.tight_layout()
        plt.savefig('distribution_before_after_outlier_removal.png',
                    dpi=300, bbox_inches='tight')
        plt.close()

    return df_clean, outlier_summary_df


# Remove outliers from numerical columns
data_clean, outlier_summary = remove_outliers(data, numerical_columns)

# 5. Normalize numerical features using Min-Max scaling
scaler = MinMaxScaler()
data_clean[numerical_columns] = scaler.fit_transform(
    data_clean[numerical_columns])

# 6. Generate enhanced correlation heatmap


def plot_enhanced_correlation_heatmap(data, output_file='enhanced_correlation_heatmap.png'):
    """
    Creates an enhanced correlation heatmap with better visualization.

    Parameters:
    - data: DataFrame containing the features
    - output_file: Path to save the output figure
    """
    plt.figure(figsize=(18, 16))

    # Compute correlation matrix
    correlation_matrix = data.corr()

    # Create mask for the upper triangle
    mask = np.triu(correlation_matrix)

    # Create a diverging colormap
    cmap = sns.diverging_palette(230, 20, as_cmap=True)

    # Enhanced heatmap
    heatmap = sns.heatmap(
        correlation_matrix,
        mask=mask,
        annot=True,
        fmt='.2f',
        cmap=cmap,
        vmin=-1,
        vmax=1,
        center=0,
        square=True,
        linewidths=.5,
        cbar_kws={"shrink": .8, "label": "Correlation Coefficient"},
        annot_kws={"size": 7}
    )

    # Add improved title and labels
    plt.title('Feature Correlation Coefficient Heatmap', fontsize=16, pad=20)

    # Improve axis labels visibility
    plt.xticks(rotation=45, ha='right', fontsize=9)
    plt.yticks(fontsize=9)

    # Add text explaining the interpretation
    plt.figtext(0.5, 0.01,
                "Interpretation: 1.0 = Perfect positive correlation, -1.0 = Perfect negative correlation, 0 = No correlation",
                ha="center", fontsize=10, bbox={"facecolor": "lightgrey", "alpha": 0.5, "pad": 5})

    # Highlight strong correlations with annotation
    rows, cols = np.where((np.abs(correlation_matrix) > 0.7)
                          & (np.abs(correlation_matrix) < 1.0))
    for i, j in zip(rows, cols):
        if i < j:  # Only annotate upper triangle
            continue
        plt.annotate(
            'Strong',
            xy=(j, i),
            xytext=(j+0.3, i+0.3),
            arrowprops=dict(arrowstyle='->', color='black', lw=0.5),
            fontsize=7,
            color='navy',
            backgroundcolor='white'
        )

    plt.tight_layout()
    plt.savefig(output_file, dpi=300, bbox_inches='tight')
    plt.close()

    # Additionally create a tabular view of strongest correlations
    high_corr = correlation_matrix.unstack().sort_values(ascending=False)
    # Drop self-correlations (which are always 1.0)
    high_corr = high_corr[high_corr < 1.0]
    top_correlations = high_corr.head(15)

    # Create a figure for top correlations
    plt.figure(figsize=(10, 6))

    # Convert to DataFrame for better visualization
    top_corr_df = pd.DataFrame({
        'Feature Pair': [f"{idx[0]}  {idx[1]}" for idx in top_correlations.index],
        'Correlation': top_correlations.values
    })

    # Plot as horizontal bar chart
    ax = sns.barplot(x='Correlation', y='Feature Pair',
                     data=top_corr_df, palette='viridis')

    # Add value labels
    for i, v in enumerate(top_corr_df['Correlation']):
        ax.text(v + 0.01, i, f'{v:.3f}', va='center')

    plt.title('Top 15 Strongest Feature Correlations', fontsize=14)
    plt.xlabel('Correlation Coefficient')
    plt.ylabel('Feature Pair')
    plt.grid(axis='x', linestyle='--', alpha=0.7)
    plt.tight_layout()
    plt.savefig('top_correlations.png', dpi=300, bbox_inches='tight')
    plt.close()


# Generate enhanced correlation heatmap
plot_enhanced_correlation_heatmap(
    data_clean, 'enhanced_correlation_heatmap.png')

# 7. Random Forest Feature Selection with improved visualization


def enhanced_random_forest_feature_selection(data, target_column, n_estimators=100, n_features_to_select=10):
    """
    Enhanced feature selection using Random Forest with improved visualization.

    Parameters:
    - data: DataFrame containing the features and target
    - target_column: Name of the target column to predict
    - n_estimators: Number of trees in the Random Forest
    - n_features_to_select: Number of top features to select

    Returns:
    - selected_features: List of selected feature names
    - X_selected: DataFrame containing only the selected features
    """
    # Prepare data
    X = data.drop(target_column, axis=1)
    y = data[target_column]

    print(f"\nFeature Selection Analysis for Target: {target_column}")
    print(f"Total features available: {X.shape[1]}")

    # Train Random Forest model
    rf_model = RandomForestRegressor(
        n_estimators=n_estimators,
        random_state=42,
        n_jobs=-1,
        max_depth=None,
        min_samples_split=2,
        min_samples_leaf=1
    )

    # Fit the model
    rf_model.fit(X, y)

    # Get feature importances
    feature_importances = pd.DataFrame({
        'Feature': X.columns,
        'Importance': rf_model.feature_importances_
    }).sort_values('Importance', ascending=False)

    # Calculate cumulative importance
    feature_importances['Cumulative_Importance'] = feature_importances['Importance'].cumsum()

    # Create enhanced horizontal bar chart for feature importance
    plt.figure(figsize=(12, max(8, min(15, len(feature_importances) * 0.3))))

    # Use a colormap that gradually changes color based on importance
    colors = plt.cm.viridis(np.linspace(
        0, 1, len(feature_importances.head(20))))

    # Plot horizontal bars
    bars = plt.barh(
        y=np.arange(len(feature_importances.head(20))),
        width=feature_importances['Importance'].head(20),
        color=colors,
        edgecolor='grey',
        alpha=0.8
    )

    # Set y-ticks to feature names
    plt.yticks(np.arange(len(feature_importances.head(20))),
               feature_importances['Feature'].head(20))

    # Add importance values as text
    for i, (importance, feature) in enumerate(zip(feature_importances['Importance'].head(20),
                                                  feature_importances['Feature'].head(20))):
        plt.text(importance + 0.005, i,
                 f'{importance:.4f}', va='center', fontsize=9)

    # Add threshold line for feature selection
    threshold = feature_importances.iloc[n_features_to_select-1]['Importance']
    plt.axvline(x=threshold, color='red', linestyle='--',
                label=f'Selection Threshold\n(Top {n_features_to_select} Features)')

    # Customize the plot
    plt.title(
        f'Random Forest Feature Importance for {target_column}', fontsize=15, pad=20)
    plt.xlabel('Importance Score (Higher = More Predictive)', fontsize=12)
    plt.ylabel('Features', fontsize=12)
    plt.legend(loc='lower right')
    plt.grid(axis='x', linestyle='--', alpha=0.6)

    # Add annotations explaining feature importance
    plt.figtext(0.5, 0.01,
                "Feature importance indicates how useful each feature is in predicting the target variable.\n"
                "Higher values mean the feature has greater influence on the model's predictions.",
                ha="center", fontsize=10, bbox={"facecolor": "lightyellow", "alpha": 0.5, "pad": 5})

    plt.tight_layout(rect=[0, 0.03, 1, 0.97])
    plt.savefig(f'feature_importance_{target_column.replace(" ", "_").replace("(", "").replace(")", "").replace("%", "pct")}.png',
                dpi=300, bbox_inches='tight')

    # Create cumulative importance plot
    plt.figure(figsize=(12, 7))

    # Plot cumulative importance curve
    plt.plot(range(1, len(feature_importances) + 1),
             feature_importances['Cumulative_Importance'],
             marker='o', linestyle='-', color='blue', linewidth=2,
             label='Cumulative Importance')

    # Add reference lines
    plt.axhline(y=0.95, color='red', linestyle='--',
                label='95% Importance Threshold')

    # Find how many features needed for 95% importance
    features_for_95 = sum(feature_importances['Cumulative_Importance'] <= 0.95)
    plt.axvline(x=features_for_95 + 1, color='green', linestyle='--',
                label=f'{features_for_95 + 1} Features needed for 95% Importance')

    # Highlight our selection
    plt.axvline(x=n_features_to_select, color='purple', linestyle='-.',
                label=f'Selected Top {n_features_to_select} Features')

    # Customize the plot
    plt.title(
        f'Cumulative Feature Importance for {target_column}', fontsize=15, pad=20)
    plt.xlabel('Number of Features', fontsize=12)
    plt.ylabel('Cumulative Importance', fontsize=12)
    plt.xticks(range(0, len(feature_importances) + 1, 5))
    plt.yticks(np.arange(0, 1.1, 0.1))
    plt.grid(True, linestyle='--', alpha=0.6)
    plt.legend(loc='lower right')

    # Add explanation text
    plt.figtext(0.5, 0.01,
                "This plot shows how many features are needed to capture a certain percentage of the total predictive power.\n"
                "Fewer features needed to reach high cumulative importance means the model relies heavily on those key features.",
                ha="center", fontsize=10, bbox={"facecolor": "lightyellow", "alpha": 0.5, "pad": 5})

    plt.tight_layout(rect=[0, 0.03, 1, 0.97])
    plt.savefig(f'cumulative_importance_{target_column.replace(" ", "_").replace("(", "").replace(")", "").replace("%", "pct")}.png',
                dpi=300, bbox_inches='tight')
    plt.close()

    # Select top N features
    top_features = feature_importances['Feature'].head(
        n_features_to_select).tolist()
    X_selected = X[top_features]

    # Print feature selection results
    print(
        f"\nSelected top {n_features_to_select} features for predicting {target_column}:")
    print(f"Total importance captured: {feature_importances['Importance'].head(n_features_to_select).sum():.4f} " +
          f"({feature_importances['Importance'].head(n_features_to_select).sum() * 100:.2f}% of total)")

    for i, (feature, importance) in enumerate(zip(
            feature_importances['Feature'].head(n_features_to_select),
            feature_importances['Importance'].head(n_features_to_select))):
        print(f"{i+1}. {feature}: {importance:.4f} ({importance * 100:.2f}%)")

    # Calculate how many features needed for various importance thresholds
    for threshold in [0.75, 0.9, 0.95, 0.99]:
        features_needed = sum(
            feature_importances['Cumulative_Importance'] <= threshold) + 1
        print(
            f"Features needed for {threshold*100:.0f}% importance: {features_needed}")

    return top_features, X_selected


# 7. Perform feature selection for the main target
target_column = 'Neural Link Stability (%)'
top_features, X_selected = enhanced_random_forest_feature_selection(
    data_clean,
    target_column=target_column,
    n_estimators=100,
    n_features_to_select=10
)

# Print selected dataset information
print(
    f"\nFeature selection complete. Selected dataset shape: {X_selected.shape}")

# 7.1 Optionally, try different targets for feature selection
alternative_targets = ['Sync Stability (%)', 'Quest Exploit Score']
for alt_target in alternative_targets:
    print(f"\n\n{'='*80}")
    print(f"Alternative Target Analysis: {alt_target}")
    print(f"{'='*80}")
    alt_features, _ = enhanced_random_forest_feature_selection(
        data_clean,
        target_column=alt_target,
        n_estimators=100,
        n_features_to_select=5  # Fewer features for alternative targets
    )

# 8. Generate summary report with key findings


def generate_summary_report(data_original, data_clean, outlier_summary, top_features, target_column):
    """
    Generates a comprehensive summary of the data preprocessing and feature selection.
    """
    print("\n" + "="*80)
    print(" "*30 + "ANALYSIS SUMMARY REPORT")
    print("="*80)

    # 1. Dataset Overview
    print("\n1. DATASET OVERVIEW")
    print(
        f"   Original dataset shape: {data_original.shape} ({data_original.shape[0]} rows, {data_original.shape[1]} columns)")
    print(
        f"   Cleaned dataset shape: {data_clean.shape} ({data_clean.shape[0]} rows, {data_clean.shape[1]} columns)")
    print(
        f"   Rows removed: {data_original.shape[0] - data_clean.shape[0]} ({(data_original.shape[0] - data_clean.shape[0])/data_original.shape[0]*100:.2f}%)")

    # 2. Outlier Summary
    print("\n2. OUTLIER SUMMARY")
    top_outlier_columns = outlier_summary.sort_values(
        'Outlier_Count', ascending=False).head(5)
    print("   Top 5 columns with most outliers:")
    for _, row in top_outlier_columns.iterrows():
        print(
            f"   - {row['Column']}: {row['Outlier_Count']} outliers ({row['Outlier_Percentage']:.2f}%)")

    # 3. Feature Selection Summary
    print(f"\n3. FEATURE SELECTION SUMMARY (Target: {target_column})")
    print(
        f"   Selected {len(top_features)} out of {data_clean.shape[1]-1} available features")
    print("   Top 5 features by importance:")
    for i, feature in enumerate(top_features[:5]):
        print(f"   {i+1}. {feature}")

    # 4. Correlation Summary
    print("\n4. CORRELATION SUMMARY")
    # Get correlation matrix and find highest correlations
    corr_matrix = data_clean.corr()
    # Get the upper triangle of the correlation matrix
    upper = corr_matrix.where(
        np.triu(np.ones(corr_matrix.shape), k=1).astype(bool))

    # Find the top 5 highest correlations
    highest_corrs = upper.unstack().sort_values(ascending=False)[:5]
    print("   Top 5 highest feature correlations:")
    for (col1, col2), value in highest_corrs.items():
        print(f"   - {col1} ~ {col2}: {value:.3f}")

    # 5. Recommendations
    print("\n5. RECOMMENDATIONS")

    # Check for highly correlated features in top_features
    top_features_corr = corr_matrix.loc[top_features, top_features]
    high_corr_pairs = []
    for i in range(len(top_features)):
        for j in range(i+1, len(top_features)):
            if abs(top_features_corr.iloc[i, j]) > 0.7:
                high_corr_pairs.append(
                    (top_features[i], top_features[j], top_features_corr.iloc[i, j]))

    if high_corr_pairs:
        print(
            "   - Consider removing some highly correlated features from the selected set:")
        for feature1, feature2, corr in high_corr_pairs:
            print(
                f"     * {feature1} and {feature2} (correlation: {corr:.3f})")

    # Check for features with many outliers
    high_outlier_features = outlier_summary[outlier_summary['Outlier_Percentage'] > 5]['Column'].tolist(
    )
    if high_outlier_features:
        print(
            "   - Consider different handling for features with high outlier percentages:")
        for feature in high_outlier_features:
            print(f"     * {feature}")

    print("\n" + "="*80)


# Generate comprehensive summary report
generate_summary_report(data_with_categories, data_clean,
                        outlier_summary, top_features, target_column)

print("\nPipeline execution complete! All visualizations have been saved.")
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import LabelEncoder, MinMaxScaler
from sklearn.ensemble import RandomForestRegressor
import warnings
warnings.filterwarnings('ignore')

# Set random seed
np.random.seed(42)

# Load data
data = pd.read_csv('updated_neoverse_logs.csv')

# Process timestamps
data['Timestamps'] = pd.to_datetime(data['Timestamps'])
data['Day'] = data['Timestamps'].dt.day
data['Month'] = data['Timestamps'].dt.month
data['Hour'] = data['Timestamps'].dt.hour
data['Weekday'] = data['Timestamps'].dt.weekday
data = data.drop('Timestamps', axis=1)

# Define column types
categorical_columns = ['Player ID', 'Player Rank', 'Team Affiliation', 'VIP Status',
                       'Player Level', 'Dark Market Transactions']
numerical_columns = ['Hours Played', 'Money Spent ($)', 'Criminal Score', 'Missions Completed',
                     'Cash on Hand ($)', 'Sync Stability (%)', 'Quest Exploit Score',
                     'Transaction Amount ($)', 'Neural Link Stability (%)',
                     'Day', 'Month', 'Hour', 'Weekday']

# Label encoding
for column in categorical_columns:
    le = LabelEncoder()
    data[f'{column}_encoded'] = le.fit_transform(data[column])

# Store original data and drop categorical columns
data_original = data.copy()
data = data.drop(categorical_columns, axis=1)
# Drop Player ID encoded column as well
if 'Player ID_encoded' in data.columns:
    data = data.drop('Player ID_encoded', axis=1)
    print("Dropped 'Player ID_encoded' column")
# Plot individual boxplots for outlier detection
for column in numerical_columns:
    # Convert pixels to inches (assuming 100 DPI)
    plt.figure(figsize=(450/100, 400/100))
    sns.boxplot(x=data[column], orient='h')

    # Add IQR bounds
    Q1 = data[column].quantile(0.25)
    Q3 = data[column].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR

    plt.axvline(x=lower_bound, color='red', linestyle='--')
    plt.axvline(x=upper_bound, color='red', linestyle='--')

    # Add title with outlier count
    outlier_count = len(
        data[(data[column] < lower_bound) | (data[column] > upper_bound)])
    plt.title(f'{column}')

    plt.tight_layout()
    plt.savefig(
        f'boxplot_{column.replace(" ", "_").replace("(", "").replace(")", "").replace("%", "pct")}.png', dpi=100)
    plt.close()

# Remove outliers


def remove_outliers(df, columns):
    df_clean = df.copy()
    outlier_indices = []

    for column in columns:
        Q1 = df[column].quantile(0.25)
        Q3 = df[column].quantile(0.75)
        IQR = Q3 - Q1
        lower_bound = Q1 - 1.5 * IQR
        upper_bound = Q3 + 1.5 * IQR

        outliers = df[(df[column] < lower_bound) |
                      (df[column] > upper_bound)].index
        outlier_indices.extend(outliers)

    outlier_indices = list(set(outlier_indices))
    df_clean = df_clean.drop(outlier_indices)

    print(
        f"Removed {len(outlier_indices)} rows containing outliers ({len(outlier_indices)/len(df)*100:.2f}%)")
    return df_clean


data_clean = remove_outliers(data, numerical_columns)

# Scale numerical features
scaler = MinMaxScaler()
data_clean[numerical_columns] = scaler.fit_transform(
    data_clean[numerical_columns])

# Generate correlation heatmap
plt.figure(figsize=(16, 14))
correlation_matrix = data_clean.corr()
mask = np.triu(correlation_matrix)
cmap = sns.diverging_palette(220, 10, as_cmap=True)

sns.heatmap(
    correlation_matrix,
    annot=True,
    fmt=".2f",
    cmap=cmap,
    mask=mask,
    vmin=-1, vmax=1
)

plt.title("Feature Correlation Coefficient Heatmap")
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.savefig('correlation_heatmap.png', dpi=300)
plt.close()

# Random Forest Feature Selection


def rf_feature_selection(data, target_column, n_features=10):
    X = data.drop(target_column, axis=1)
    y = data[target_column]

    # Train model
    rf_model = RandomForestRegressor(
        n_estimators=100, random_state=42, n_jobs=-1)
    rf_model.fit(X, y)

    # Get feature importances
    feature_importances = pd.DataFrame({
        'Feature': X.columns,
        'Importance': rf_model.feature_importances_
    }).sort_values('Importance', ascending=False)

    # Plot feature importances
    plt.figure(figsize=(14, 10))
    ax = sns.barplot(x='Importance', y='Feature',
                     data=feature_importances.head(20))

    # Add importance values
    for i, importance in enumerate(feature_importances['Importance'].head(20)):
        ax.text(importance + 0.005, i, f'{importance:.4f}', va='center')

    plt.title(f'Feature Importance for {target_column}')
    plt.xlabel('Importance Score')
    plt.ylabel('Features')
    plt.tight_layout()
    plt.savefig(
        f'feature_importance_{target_column.replace(" ", "_").replace("(", "").replace(")", "").replace("%", "pct")}.png')
    plt.close()

    # Select top features
    top_features = feature_importances['Feature'].head(n_features).tolist()
    return top_features


# Perform feature selection for main target
target_column = 'Neural Link Stability (%)'
top_features = rf_feature_selection(data_clean, target_column, 10)
print(f"\nSelected top 10 features for {target_column}:")
for i, feature in enumerate(top_features):
    print(f"{i+1}. {feature}")

# Try different targets
alternative_targets = ['Dark Market Transactions_encoded', 'Sync Stability (%)',
                       'Quest Exploit Score', 'Transaction Amount ($)', 'Criminal Score']
for alt_target in alternative_targets:
    print(f"\nFeature selection for: {alt_target}")
    alt_features = rf_feature_selection(data_clean, alt_target, 5)
    print(f"Top 5 features for {alt_target}:")
    for i, feature in enumerate(alt_features):
        print(f"{i+1}. {feature}")

print("\nAnalysis complete! All visualizations have been saved.")
```
